## Mencari Sudut untuk Accelerometer dan Gyroscope

### Cari Sudut Accelerometer
image.png

```python
#Membuat fungsi rumus untuk mencari sudut acceleration
def sudutAccel (x,y,z):
  isian = z/math.sqrt((y*y)+(x*x))
  sudut = math.atan(isian)
  return sudut

dataSudutAccel= {"Sudut Accelerometer": []}  
df = pd.DataFrame(dataSudutAccel)

lengtha = len(x_accel)
i = 0
for i in range (lengtha):
  values = 180/pi * sudutAccel(x_accel[i], y_accel[i], z_accel[i])/2
  df.loc[i] = [values]

dataSudutAccel = df
print("Sudut Max, Sudut Min :")
print(dataSudutAccel.max())
print(dataSudutAccel.min())
print(dataSudutAccel)
```

```python
#Menampilkan hasil Plot
plt.figure(figsize=(20,5))
plt.title('Sudut Accelerometer')
plt.xlabel('Time (s)')
plt.ylabel('Derajat')
plt.plot(time_accel, dataSudutAccel)
plt.show
```

### Cari Sudut Gyroscope
image.png

```python
def gyroAngle(Dz_gyro, time_gyro, gtimePrevious):
  kenaikanWaktu = time_gyro - gtimePrevious
  kenaikanSudut = Dz_gyro * kenaikanWaktu
  gyroAngle = (Dz_gyro + kenaikanSudut)/2
  return gyroAngle
  
dataSudutGyro = {"Sudut Gyro": []}  
df = pd.DataFrame(dataSudutGyro)

lengthg = len(Dz_gyro)
i = 0
for i in range (lengthg):
  if (i > 0):
    values = 180 / pi* (gyroAngle(y_gyro[i], time_gyro[i], time_gyro[i-1]))/2
    df.loc[i] = [values]
  else:
    df.loc[i] = y_gyro[i]
  
dataSudutGyro = df

plt.figure(figsize=(10, 5))
plt.title('Sudut Gyro')
plt.xlabel('Time (s)')
plt.ylabel('Derajat')
plt.plot(time_gyro, dataSudutGyro)
plt.show()
```

## CF (Complementary Filter)

image.png

```python
# Membuat fungsi untuk membuat Rumus Complementary Filter
def complementaryFilter(prevSudut ,dataSudutGyro,dataSudutAccel):
  sudut = 0.98 * (dataSudutGyro) + 0.02 * (dataSudutAccel)/2
  return sudut

dataSudutCF = {"Sudut Complementary Filter": []}  
df_CF = pd.DataFrame(dataSudutCF)


#Jumlah Sampel Accelerometer dan Gyrometer saya ternyata berbeda sehingga saya pakai 10000 data saja
panjangSampel = 10000


prevSudut = 0
for i in range (panjangSampel):
  values = complementaryFilter(prevSudut, dataSudutGyro['Sudut Gyro'][i], dataSudutAccel['Sudut Accelerometer'][i])
  df_CF.loc[i] = [values]
  prevSudut = values

dataSudutCF = df_CF
```

```python
timeCF = {"time CF": []}  
df_CF = pd.DataFrame(timeCF)
for i in range (10000):
  df_CF.loc[i] = time_gyro[i]
timeCF = df_CF

plt.figure(figsize=(25,5))
plt.title('Sudut hasil CF')
plt.ylabel('derajat')
plt.xlabel('Time')
plt.plot(timeCF,dataSudutCF)
print(dataSudutCF.max())
print(dataSudutCF.min())
plt.show
```

### Sudut Gait Angle

```python
dataSudutGait = {"Sudut Gait": []}  
df_GaitAngle = pd.DataFrame(dataSudutGait)

def gaitAngle(dataSudutCF):
  if (dataSudutCF < 0 ):
    return 0
  else:
    return dataSudutCF

for i in range (panjangSampel):
  values = gaitAngle(dataSudutCF['Sudut Complementary Filter'][i])
  df_GaitAngle.loc[i] = [values]

dataSudutGait = df_GaitAngle
```

```python
plt.figure(figsize=(25,6))
plt.title('Sudut Gait Lutut Kaki Kanan')
plt.xlabel('Time')
plt.ylabel('derajat')
plt.plot(timeCF,dataSudutGait)
plt.show()
```
